# 研究人员发现了大规模大脑模拟的算法
> [原文链接](https://blog.frontiersin.org/2018/03/02/neuroscience-brain-simulation-algorithm-exascale/?utm_source=G-BLO&utm_medium=WTXT&utm_campaign=ECO_FPSYG_20180427_AI-optical-illusion)

人类大脑是一个复杂的难以置信的器官，由1000亿个相互连接的神经元组成。 然而，即使在最强大的超级计算机的帮助下，目前也不可能模拟这种大小的网络中神经元信号的交换。

一个国际研究小组现在已经朝着创造实现大脑规模网络模拟百万兆级超级计算机的技术迈出了决定性的一步。 这个突破性的算法，发表在《Frontiers in Neuroinformatics》，该算法实现了使用相同数量的计算机内存来表示人脑中较大的区域。 同时，新算法显著加速了现有超级计算机的大脑模拟。

"自2014年以来，我们的软件可以对人脑中大约百分之一的神经元与它们的所有连接进行模拟,"于利希神经科学与医学研究所主任马库斯 · 迪斯曼说。 为了实现这一令人印象深刻的壮举，该软件需要整个千兆级超级计算机的主要内存，比如日本神户的 超级计算机**K** 和德国于利希的 **JUQUEEN**。

> 千兆级计算是指一台计算机的峰值计算速度可以达到万亿次每秒。即一个PFLOPS（petaFLOPS）(文中称千兆)等于每秒一千兆（$10^{15}$ ）次的浮点运算。一个EFLOPS（exaFLOPS）（文中称百万兆）等于每秒一佰京（$10^{18}$）次的浮点运算。

> 在截止2017年11月发稿时的的[超级计算机列表](https://www.top500.org/list/2017/11/?page=1)中，超算[JUQUEEN](https://www.top500.org/system/177232)排名第22，[K](https://www.top500.org/system/177232)排名第10。详情请点超链接。

柴尔曼已经为模拟软件NEST工作了20多年 —— NEST是一个被神经科学界广泛使用的免费开源模拟代码，也是**欧洲人脑计划**的核心模拟器。在欧洲人脑计划中，柴尔曼领导着理论神经科学和高性能分析与计算平台等领域的项目。

在NEST中，网络中每个神经元的行为都是由一些数学方程组来表示的。未来的百万兆级超级计算机，比在在建的神户超算 post-K 和于利希的 JUWELS，将是如今高端超级计算机新能的10到100倍。 研究人员将首次拥有计算能力来模拟人类大脑规模的神经网络。

Seemingly a dead end

## 似乎是个死胡同

虽然目前的模拟技术使研究人员能够开始研究大型神经元网络，但它也代表了发展百万兆级技术的死胡同。 超级计算机由大约10万台小型计算机组成，这些计算机被称为节点，每台计算机都配备了一些处理器来进行实际的计算。

"在神经元网络模拟开始之前，神经元及其连接需要被创建出来，这意味着它们需要在节点的内存中实例化。 在模拟过程中，一个神经元不知道它的目标神经元在哪个节点运行，因此，它的短电脉冲需要发送到所有的节点。 每个节点然后检查所有这些电脉冲，看哪一个与这个节点上存在的虚拟神经元相关,"位于斯德哥尔摩的 KTH 皇家理工学院的 Susanne Kunkel 解释道。

目前的网络创建算法是高效的，因为所有节点同时构造网络的属于本节点的特定部分。 然而，将所有电脉冲发送到所有节点并不适合于在百万兆级系统上的模拟。

"要高效地检查每个电脉冲的相关性，需要为整个网络中的每个神经元中的每个处理器都提供一比特（一个二进制位）的信息。 对于一个拥有10亿个神经元的网络，每个节点中的大部分内存都被表示神经元的这部分信息消耗掉了。"。

这是模拟更大网络时的主要问题: 没个节点上的内存需要随着整个神经网络的神经元的数量而增加。 从人类大脑的尺度来看，这就要求每个处理器的可用内存要比现在的超级计算机大100倍。 然而，下一代超级计算机的情况不太可能如此。 每个计算节点的处理器数量将会增加，但是每个处理器的内存和计算节点的数量将保持不变。
