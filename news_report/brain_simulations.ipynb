# 研究人员发现了大规模大脑模拟的算法
> [原文链接](https://blog.frontiersin.org/2018/03/02/neuroscience-brain-simulation-algorithm-exascale/?utm_source=G-BLO&utm_medium=WTXT&utm_campaign=ECO_FPSYG_20180427_AI-optical-illusion)

![img](https://frontiersinblog.files.wordpress.com/2018/03/neuroscience-brain-simulation-algorithm-exascale.jpg)
这一突破性进展是构建超级计算机神经网络的一种新方法。

人类大脑是一个复杂的难以置信的器官，由1000亿个相互连接的神经元组成。 然而，即使在最强大的超级计算机的帮助下，目前也不可能模拟这种大小的网络中神经元信号的交换。

一个国际研究小组现在已经朝着创造实现大脑规模网络模拟百万兆级超级计算机的技术迈出了决定性的一步。 这个突破性的算法，发表在《Frontiers in Neuroinformatics》，该算法实现了使用相同数量的计算机内存来表示人脑中较大的区域。 同时，新算法显著加速了现有超级计算机的大脑模拟。

"自2014年以来，我们的软件可以对人脑中大约百分之一的神经元与它们的所有连接进行模拟,"于利希神经科学与医学研究所主任Markus Diesmann说。 为了实现这一令人印象深刻的壮举，该软件需要整个千兆级超级计算机的主要内存，比如日本神户的 超级计算机**K** 和德国于利希的 **JUQUEEN**。

> 千兆级计算是指一台计算机的峰值计算速度可以达到万亿次每秒。即一个PFLOPS（petaFLOPS）(文中称千兆)等于每秒一千兆（$10^{15}$ ）次的浮点运算。一个EFLOPS（exaFLOPS）（文中称百万兆）等于每秒一佰京（$10^{18}$）次的浮点运算。

> 在截止2017年11月发稿时的的[超级计算机列表](https://www.top500.org/list/2017/11/?page=1)中，超算[JUQUEEN](https://www.top500.org/system/177232)排名第22，[K](https://www.top500.org/system/177232)排名第10。详情请点超链接。

Diesmann 已经为模拟软件 NEST 工作了20多年 —— NEST是一个被神经科学界广泛使用的免费开源模拟代码，也是**欧洲人脑计划**的核心模拟器。在欧洲人脑计划中，他领导着理论神经科学和高性能分析与计算平台等领域的项目。

在NEST中，网络中每个神经元的行为都是由一些数学方程组来表示的。未来的百万兆级超级计算机，如在建的神户超算 post-K 和于利希的 JUWELS，将是如今高端超级计算机新能的10到100倍。 研究人员将首次拥有计算能力来模拟人类大脑规模的神经网络。

## 山重水复疑无路

虽然目前的模拟技术使研究人员能够开始研究大型神经元网络，但它也代表了发展百万兆级技术的瓶颈。超级计算机由大约10万台小型计算机组成，这些计算机被称为节点，每台计算机都配备了一些处理器来进行实际的计算。

"在神经元网络模拟开始之前，神经元及其连接需要被创建出来，这意味着它们需要在节点的内存中实例化。 在模拟过程中，一个神经元不知道它的目标神经元在哪个节点运行，因此，它的短电脉冲需要发送给所有的节点。 然后，每个节点检查所有电脉冲，看哪一个节点与这个节点上存在的虚拟神经元相关。"位于斯德哥尔摩的 KTH 皇家理工学院Susanne Kunkel 解释道。

目前网络创建算法是高效的，因为所有节点同时构造网络中属于本节点的特定部分。 然而，将所有电脉冲发送到所有节点并不适合于在百万兆级系统上的模拟。

“要高效地检查每个电脉冲的相关性，需要为整个网络中的每一神经元中的每个处理器都提供1比特（一个二进制位）的信息。 对于一个拥有10亿神经元的网络，每个节点中的大部分内存都被表示神经元的这部分信息消耗掉了。” Markus Diesmann 补充说明。

这是模拟更大网络时的主要问题: 每个节点上的内存需要随着整个神经网络的神经元的数量而增加。 从人类大脑的规模来看，这就要求每个处理器的可用内存要比现在的超级计算机大100倍。 然而，下一代超级计算机的情况不太可能如此。 每个计算节点的处理器数量将会增加，但是每个处理器的内存和计算节点的数量将保持不变。

## 柳暗花明又一村

这一突破发表在《Frontiers in Neuroinformatics》中，是构建超级计算机神经网络的一种新方法。根据本算法，每个节点上所需的内存不再随网络大小增加。 在模拟开始时，新技术允许节点交换关于谁需要向谁发送神经元活动数据的信息。 一旦这些信息提供的知识是有效的，节点之间的神经元活动数据的交换行为被重构，每个节点只接收它所需要的信息。 网络中的每个神经元不再需要额外的比特。

## 踏破铁鞋无觅处

在测试他们的新想法时，科学家们获得一个额外的关键洞见，Susanne Kunkel 说:"当分析新的算法时，我们意识到我们的新技术不仅能够对超级计算机系统进行仿真，而且还能使模拟在现有超级计算机上更快。"

事实上，随着内存消耗的控制，模拟的速度成为进一步技术发展的主要焦点。 例如，在于利希的超级计算机 JUQUEEN 上，一个由5.8万亿突触连接的0.52亿个神经元的大型模拟，需要28.5分钟来计算生物时间的1秒钟。 通过改进仿真的数据结构，时间缩短到5.2分钟。

“有了新技术，我们可以比以前更好地利用现代微处理器不断增加的并行性，这在百万兆级超级计算机领域将变得更加重要。”该研究报告的主要作者 Jakob Jordan 表示。

“百万兆级的硬件和适当软件的结合，使我们对大脑功能的基本方面进行了研究，比如我们能够获得的可塑性和学习能力.” Markus Diesmann 补充道。

在模拟软件NEST的下一次发布中，研究人员会将他们的成果开源，免费提供给社区。

“我们一直在超级计算机 K 上，使用 NEST 来模拟健康和帕金森氏病的基底神经节回路的复杂动态。我们很高兴听到有关新一代 NEST 的消息，这将使我们能在超级计算机 post-K 上进行全脑模拟，以弄清楚运动控制和心理功能的神经机制。” OIST的Kenji Doya 说。

“这项研究是国际合作建造百万兆级超级计算机的一个很好的例子。重要的是，我们已经准备好了应用程序，随时等待着这些宝贵的机器投入使用。”神户的 RIKEN 计算机科学高级研究所 Mitsuhisa Sato 总结道。
